```python
# data
from keras.datasets import mnist
from keras.datasets import boston_housing
# packages
import matplotlib.pyplot as plt
import numpy as np
import random
from keras import models
from keras.models import Sequential
from keras.layers import Dense
from keras import layers
from keras.layers import Dropout
from keras.utils import np_utils
from keras.regularizers import l1, l2
from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
import random
random.seed(1234)
```

## Part 1

### 1-1 Initial test

```python
# Load the MNIST dataset(train, test)
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 2D tensor with individual values between 0 and 1 (prepare the image data)
train_images = train_images.reshape( (60000, 28 * 28) )
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape( (10000, 28 * 28) )
test_images = test_images.astype('float32') / 255

# prepare labels
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
# create five layers
network1 = models.Sequential()
network1.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network1.add(layers.Dense(512, activation='relu'))
network1.add(layers.Dense(512, activation='relu'))
network1.add(layers.Dense(512, activation='relu'))
network1.add(layers.Dense(10, activation='softmax'))

# compile and train
network1.compile(optimizer='rmsprop', loss='categorical_crossentropy', \
                metrics=['accuracy'])
m1 = network1.fit(train_images, train_labels, epochs=200, \
                  batch_size=512, validation_split= 1/6)
```

```python
# plot the validation set accuracy and loss over the epochs
plt.plot(m1.history['acc'])
plt.plot(m1.history['val_acc'])
plt.title('Figure1 Validation Accuracy Over the Epochs')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'])
plt.show()

plt.plot(m1.history['loss'])
plt.plot(m1.history['val_loss'])
plt.title('Figure2 Validation Loss Over the Epochs')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'])
plt.show()
```

```python
# Identify the epoch where the model's
# performance degrades based on the validation set
m1.history['val_loss'].index(min(m1.history['val_loss']))+1
```

At around epoch 6, we have the minimum loss on validation. The degrading trend starts from epoch 7.

### 1-2 Implement dropout

```python
# drop 0.5
network2 = Sequential()
network2.add(Dense(512, kernel_initializer='normal', activation='relu',\
                input_shape=(28 * 28,)))
network2.add(Dropout(0.5))
network2.add(Dense(512, kernel_initializer='normal', activation='relu'))
network2.add(Dropout(0.5))
network2.add(Dense(512, kernel_initializer='normal', activation='relu'))
network2.add(Dropout(0.5))
network2.add(Dense(512, kernel_initializer='normal', activation='relu'))
network2.add(Dropout(0.5))
network2.add(Dense(10, kernel_initializer='normal', activation='softmax'))
```

```python
# compile and train
network2.compile(optimizer='rmsprop', loss='categorical_crossentropy', \
                       metrics=['accuracy'])
m2 = network2.fit(train_images, train_labels, epochs=200, \
                      batch_size=512, validation_split=1/6)
```

```python
# plot and compare model2 with model1
# validation Loss plot
plt.plot(m1.history['val_loss'])
plt.plot(m2.history['val_loss'])
plt.title('Figure3 Compare Validation loss (Base & Drop)')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['Base', 'Drop'])
plt.show()
```

After drop out 0.5, the new model performs better than the initial model after around epoch 4 or 5 as it has lower validation loss.

### 1-3 Weight regularization

```python
# create layers with l1 weight regularization, compile
network3 = models.Sequential()
network3.add(layers.Dense(512, activation='relu', kernel_regularizer=l1(0.001), input_shape=(28 * 28,)))
network3.add(layers.Dense(512, kernel_regularizer=l1(0.001), activation='relu'))
network3.add(layers.Dense(512, kernel_regularizer=l1(0.001), activation='relu'))
network3.add(layers.Dense(512, kernel_regularizer=l1(0.001), activation='relu'))
network3.add(layers.Dense(10, activation='softmax'))
network3.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# create layers with l1 weight regularization, compile
network4 = models.Sequential()
network4.add(layers.Dense(512, activation='relu', kernel_regularizer=l2(0.001), input_shape=(28 * 28,)))
network4.add(layers.Dense(512, kernel_regularizer=l2(0.001), activation='relu'))
network4.add(layers.Dense(512, kernel_regularizer=l2(0.001), activation='relu'))
network4.add(layers.Dense(512, kernel_regularizer=l2(0.001), activation='relu'))
network4.add(layers.Dense(10, activation='softmax'))
network4.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
# fit network3
m3  = network3.fit(train_images, train_labels,epochs=200, \
                      batch_size=512, validation_split=1/6)
```

```python
# fit network4
m4 = network4.fit(train_images, train_labels, epochs=200, \
                      batch_size=512, validation_split=1/6)
```

```python
# plot and compare all models
# validation Loss plot
plt.plot(m1.history['val_loss'])
plt.plot(m2.history['val_loss'])
plt.plot(m3.history['val_loss'])
plt.plot(m4.history['val_loss'])
plt.title('Figure4 Compare Validation Loss (All Models)')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['Base', 'Drop', 'L1 Reg', 'L2 Reg'])
plt.show()
```

### 1-4 Final model

```python
# average loss score
print('Base model', network1.evaluate(test_images, test_labels, verbose=0)[0])
print('')
print('Drop model', network2.evaluate(test_images, test_labels, verbose=0)[0])
print('')
print('l1 Reg model', network3.evaluate(test_images, test_labels, verbose=0)[0])
print('')
print('l2 Reg model', network3.evaluate(test_images, test_labels, verbose=0)[0])
```

Drop model has the smallest average loss score.

Drop model has the smallest minimum loss score.
